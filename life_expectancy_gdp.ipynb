### Univariate Analysis for 'Expectancy'

```python
# Create a figure of width 15 and height 7 with an outer color of whitesmoke
plt.figure(figsize=(15, 7), facecolor='whitesmoke')

# Create histogram with density plot
sns.distplot(df.Expectancy, rug=True, kde=False)

# Set x-axis label
plt.xlabel('Life expectancy at birth (years)')

# Set axis inner color to 'white'
plt.axes().set_facecolor('white')

# Set title
plt.title('Distribution for Life expectancy at birth', fontweight='bold')

# Show plot
plt.show()
```

## Visualize **Expecancy** and **GDP** by yearly bases

```python
# Define a function that takes a year as an argument and plot a bar chart for 'Expectancy' and 'GDP' for the Counties on yearly bases
def bar_plots(year):
    # Filter df base on 'year'
    df_year = df[df.Year == year]
    # Create a figure of width 15 and height 5 with an outer color of whitesmoke
    plt.figure(figsize=(15,5), facecolor='whitesmoke')
    # Create the first plot
    ax1 = plt.subplot(1,2,1)
    # Plot a bar chart
    sns.barplot(x='Country', y='Expectancy', data=df_year)
    # Rotate x-axis ticks
    plt.xticks(rotation=90)
    # Set y-axis label
    plt.ylabel('Life expectancy at birth (years)')
    # Set title
    ax1.set_title('Life expectancy at birth in the year {}'.format(year), fontweight='bold')
    
    # Create the second plot
    ax2 = plt.subplot(1,2,2)
    # create a bar chart
    sns.barplot(x='Country', y='GDP', data=df_year)
    # Rotate x-axis ticks
    plt.xticks(rotation=90)
    # Set y-axis label
    plt.ylabel('GDP in Trillions of U.S. Dollars')
    # Set title
    ax2.set_title('GDP in the year {}'.format(year), fontweight='bold')
    # Show plot
    return plt.show()

# Iterate through the list of 'Year' column
for item in list(df.Year.unique()):
    # Create bar plots
    bar_plots(item)
    # Clear current plot
    plt.clf()
```

### Analysis of the above figures show that,
For about **15 years** (From the year 2000 to 2015), 
- **Chile, China, Germany, Mexico and United States of America** have a consistent life expectancy at birth at about **72 - 80** years, except for **Zimbabwe**, which has a consitent life expectancy at birth at about **45 - 50** years.
- **United States of America** consistently appears to have a much higher value in terms of GDP and then followed by China.
- **Germany** and **Mexico** have relatively close GDP.
- On the figures, Chile's GDP is very small whiles Zimbabwe's GDP is hardly seen.

## Comparing averages

```python
# Drop the 'Year' datasets for df
drop_year = df.drop('Year', axis=1)

# Calculate the mean Expectancy and mean GDP for each Country, and group dataset by 'Country'
df_mean = drop_year.groupby('Country').mean().reset_index()

# Inspect df_mean
df_mean
```

```python
# Create a figure of width 10 and height 7 with an outer color of 'whitesmoke'
plt.figure(figsize=(15, 7), facecolor='whitesmoke')

# Create the first plot axes object
ax1 = plt.subplot(1,2,1)

# Create bar chart
sns.barplot(x='Country', y='Expectancy', data=df_mean)

# Rotate x-axis tick labels
plt.xticks(rotation=90)

# Set y-axis label
plt.ylabel('Life expectancy at birth (years)')

# Set title
plt.title('Average Life expectancy at birth Distribution', fontweight='bold')

# Create the second plot axes object
ax2 = plt.subplot(1,2,2)

# Create bar chart
sns.barplot(x='Country', y='GDP', data=df_mean)

# Rotate x-axis tick labels
plt.xticks(rotation=90)

# Set y-axis label
plt.ylabel('GDP in Trillions of U.S. Dollars')

# Set title
plt.title('Average GDPs Distribution', fontweight='bold')

# Show plot
plt.show()

# Clear current figure
plt.clf()
```

**On Average,** 
- Chile, China, Germany, Mexico and United States of America have a life expectancy at birth at about 70 - 80 years.
- Zimbabwe has a life expectancy at birth at about 40 - 50 years.
- United States of America has the highest GDP which is over 1.4 Trillion U.S. dollars.
- China has the second highest GDP followed by Germany, Mexico and Chile.
- Zimbabwe has the least GDP and its distribution is not seen on the figure above.

## Examine posible correlation between Life expectancy at birth and GDP 

#### Scatter plot

```python
# Clear current plot
plt.clf()

# Create a figure of width 10 and height 7 with an outer color of 'whitesmoke'
plt.figure(figsize=(16, 7), facecolor='whitesmoke')

# Create a scatter plot
sns.scatterplot(x='Expectancy', y='GDP', hue='Country', data=df)

# Set x-axis label
plt.xlabel('Life expectancy at birth (years)')

# Set y-axis label
plt.ylabel('GDP in Trillions of U.S. Dollars')

# Show plot
plt.show()
```

**The scatter plot above illustrates that,**
- Zimbabwe continue to have increase in life expectancy whiles GDP stays almost constant
- All the other countries (Chile, China, Germany, Mexico and United States) have their life expectancy increased as GDP rises

```python
# Clear current plot
plt.clf()

# Create faceted line charts of GDP by Country
graph_corr = sns.FacetGrid(df, col='Country', col_wrap=3, hue = 'Country', sharey = False, sharex = False)

# Map dataset onto many axes in a grid of rows and columns to create line charts for 'GDP'
graph_corr = (graph_corr.map(sns.scatterplot,'Expectancy','GDP').add_legend().set_axis_labels('Life expectancy at birth (years)','GDP in Trillions of U.S. Dollars'))
```

**The above shows that,**
- An increase in GDP results in an increase in life expectancy, this exhibits a positive correlation between GDP and life expectancy.

### Spearman correlation coefficient for Life expectancy at birth and GDP

```python
# Import modules
import scipy.stats
import scipy as sp
```

```python
# Calculate Spearman correlation coefficient and p-value
r, pval = sp.stats.spearmanr(df.Expectancy,df.GDP)

# Output correlation coefficient
print('Correlation coefficient: {:.2f}'.format(r))
```

- Correlation coefficient of 0.45 indicates that there is a moderate positive correlation between GDP and Life expectancy at birth.
